```python
######################################################################
##### Single Exponential Fit, two ways ###############################
######################################################################

### Fit Input Parameters #############################################################

start         = 15.0  # Set the number of minutes after the peak value to begin the fit
fit_range     = 600.0 # Set the number of minutes to be fitted
Inj_Threshold = 5.0  # The Inj start is found to be when the Rate > (Max Rate)/Inj_Threshold 
backbin       = 10.0  # Number of minutes before trigger to count background rate
######################################################################################

kr_dc = np.log(2)/109.8 # Kr83m Decay constant

# Fitting Functions
def fit1(bin_centers, data):
    import numpy as np
    from scipy.optimize import curve_fit
    kr_dc = np.log(2)/109.8
    popt, pcov = curve_fit(lambda t,A: (A*np.exp(-t*kr_dc)), \
                           bin_centers, data, p0=[1], bounds=([0], np.inf))                                  
    A = popt[0]; sig_A = pcov[0,0]
    return (A, sig_A**(0.5))

def fit2(bin_centers, data):
    import numpy as np
    from scipy.optimize import curve_fit
    popt, pcov = curve_fit(lambda t, B, tau: (B*np.exp(-t/tau)), \
                           bin_centers, data, p0=[10, 110], bounds=([0, 0], np.inf))                                  
    B = popt[0]; tau = popt[1]; sig_B = pcov[0,0]; sig_tau = pcov[1,1];
    return (B, sig_B**(0.5), tau, sig_tau**(0.5))

# Histogram of Event Rates
plt.figure(figsize=(13,4))
y, x_tmp, _ = plt.hist(((timestamp[c_Kr83m])/(1000000.0*60.0)), \
             bins=int(math.ceil(max((timestamp/(1000000.0)))/bin_width)),\
             range=[0,math.ceil(max((timestamp/(1000000.0*60.0))))], \
             weights=w2, histtype='step',color='orange')

# Locating the 'Rising Edge' of Injected Activity
peak = max(y)
for i in x_tmp:
    if y[int(i/(bin_width/60.0))] > peak/Inj_Threshold:
        fit_t0 = i
        break
        
# Determining the Fit Range Bin Numbers
low = np.argmax(y)+int((60.0/bin_width)*start)
high = low+int((60.0/bin_width)*fit_range)

background_bin = int((fit_t0-backbin)*(60.0/bin_width))
background = y[background_bin]

# Organizing the Data to be Fed to the Fitting Function
x_tmp = x_tmp[low:high+1]
y = y[low:high]; 
x = x_tmp[:-1]+(np.diff(x_tmp)/2.0)

# Performing the Best Fits
fit_tot = []; fit_Kr =[]; fit_Other=[]; Ovr_Kr=[];
c_empty = np.array(y > 0.0) # Probably not necessary (was previously > 0, but testing)
A, sig_A = fit1(x[c_empty],y[c_empty])
B, sig_B, tau, sig_tau = fit2(x[c_empty],y[c_empty])

# Building Arrays that Contain the Best Fit Curve
for i in x:
    fit_Kr.extend([A*np.exp(-i*kr_dc)])
    fit_Other.extend([B*np.exp(-i/tau)])

fit_Kr = np.array(fit_Kr); fit_Other = np.array(fit_Other);
Amp_Kr = A*np.exp(-fit_t0*kr_dc); Amp_free = B*np.exp(-fit_t0/tau)

# Plotting the Best Fit Results
plt.plot(x, fit_Kr, color='r', label="Kr83m Half Life Fixed")
plt.plot(x, fit_Other, color='b', label = "Half Life as Free Parameter")
plt.plot([fit_t0, fit_t0],[0.01, Amp_free], color='grey')
plt.plot([fit_t0, fit_t0],[0.01, Amp_Kr], color='grey')
plt.xlabel('Minutes since '+files[0])
plt.xlim(0,math.ceil(max((timestamp/(1000000.0*60.0)))))
plt.ylabel('Rate [Bq]'); plt.yscale('log') ;plt.legend(loc='lower right')
plt.title("Single Exponential Fits, Two Different Ways"); plt.grid(); 

# Plotting the Fit Residuals
plt.figure(figsize=(13,2))
plt.plot(x[c_empty], y[c_empty]-fit_Kr[c_empty], color='r', label="Kr83m Residuals")
plt.plot(x[c_empty], y[c_empty]-fit_Other[c_empty], color='b', label="Other Residuals")
plt.plot(x, 0*x, color='k')
plt.xlabel('Minutes since '+files[0]); plt.ylabel('Residuals')
plt.xlim(0,math.ceil(max((timestamp/(1000000.0*60.0)))))
plt.legend(loc="lower right"); plt.grid(); 
plt.show()

peak_unc = np.sqrt(peak*bin_width)/bin_width
back_unc = np.sqrt(background*bin_width)/bin_width

uncert = np.sqrt( (sig_B**2) + (back_unc**2) )

print("Background Bin         = "+str(round(fit_t0-backbin,1))+" min")
print("Injection Start        = "+str(round(fit_t0,1))+" min")
print("Fixed Decay Amp        = "+str(round(Amp_Kr,2))+" +/- "+str(round(sig_A,2))+" Bq")
print("Free Decay Amp         = "+str(round(Amp_free,2))+" +/- "+str(round(sig_B,2))+" Bq")
print("Free Decay Tau         = "+str(round(tau,2))+" +/- "+str(round(sig_tau,2))+" min (Half-life = "\
      +str(round(tau*math.log(2.0),1))+" +/- "+str(round(sig_tau*math.log(2.0),1))+" min)")
print("Peak Measured Activity = "+str(round(peak,2))+" +/- "+str(round(peak_unc,2))+" Bq")
print("Background Kr83m Rate  = "+str(round(background,2))+" +/- "+str(round(back_unc,2))+" Bq")
print("Injected Activity      = "+str(round(Amp_free-background,2))+" +/- "+str(round(uncert,2))+" Bq")
```
