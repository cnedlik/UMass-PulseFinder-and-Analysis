```python
######################################################################
##### Deadtime Correction ############################################
######################################################################

bin_width = 60.0 # bin width in seconds for ALL rate plots in this notebook

ts_0 = timestamp_c[0]
deadtime_bins = int(math.ceil(((timestamp_c[-1]-ts_0)/1000000)/bin_width))
deadtimefrac_t = []; livetimefrac_t = [];
w_dead = []; 
w_dead_Rn = []; 
w_dead_Po = [];

for i in range(deadtime_bins):
    c_time_bin = np.array( (bin_width*i <= ((timestamp_c-ts_0)/1000000.0)) &\
                           (((timestamp_c-ts_0)/1000000.0) < bin_width*(i+1)))
    dead = deadtime_c[c_time_bin]
    if len(dead) != 0:
        live = livetime_c[c_time_bin]
        DF = sum(dead)/(sum(dead)+sum(live))
        LF = sum(live)/(sum(dead)+sum(live))
    else:
        DF = 0; LF = 0

    deadtimefrac_t.extend([DF]); livetimefrac_t.extend([LF])
    
    for i in range(int(len(dead))):
        w_dead.append(1.0/(1.0-DF))
    for i in range(int(len(deadtime_c[c_time_bin & c_Rn220]))):
        w_dead_Rn.append(1.0/(1.0-DF))
    for i in range(int(len(deadtime_c[c_time_bin & c_Po212]))):
        w_dead_Po.append(1.0/(1.0-DF))

deadtimefrac_t = np.array(deadtimefrac_t)
livetimefrac_t = np.array(livetimefrac_t)
```

```python
###################################################################
##### Injection Fitting ###########################################
###################################################################

time_bfor = 27.0 # number of minutes from which the avg rate is sampled before injection 
time_aftr = 18.0 # number of minutes from which the avg rate is sampled after injection
mix_t     = 10.0  # number of minutes to wait before sampling avg rate after injection
mod       = 0   # number of bins before injection to reduce time_bfor range

###################################################################

y0, x0 = np.histogram(timestamp_min_c, bins=binN, range=[lwrlim,uprlim], weights=w1)
y1, x1 = np.histogram(timestamp_min_c[c_Rn220], bins=binN, range=[lwrlim,uprlim], weights=w2)

if View_Inj_only == True:
    
    bin0 = int((ts_Inj*60.0)/bin_width) # Bin corresponding to injection start
    binC = int((lwrlim*60.0)/bin_width) # Correct for offset from injection window cut 

    low_bfor  = bin0-int((60.0/bin_width)*time_bfor)-binC
    high_bfor = bin0-binC-mod
    low_aftr  = bin0+int((60.0/bin_width)*mix_t)-binC
    high_aftr = low_aftr+int((60.0/bin_width)*time_aftr)

    x_bfor = x1[low_bfor:high_bfor+1]; x_bfor = x_bfor[:-1]+(np.diff(x_bfor)/2.0)
    x_aftr = x1[low_aftr:high_aftr+1]; x_aftr = x_aftr[:-1]+(np.diff(x_aftr)/2.0)
    y_bfor = y1[low_bfor:high_bfor];   y_aftr = y1[low_aftr:high_aftr]

    # This cut allow us to ignore the bins that have decreased rate between acquisitions in the mean
    frac = 0.0
    c_aftr = np.array(y_aftr > frac*max(y_aftr))
    print("Cutting Bins in (After) Rate Average < "+str(round(frac*max(y_aftr),1))+" Bq")
    
    mean_bfor = sum(y_bfor)/len(y_bfor)
    mean_bfor_err = np.sqrt(sum(y_bfor)*bin_width)/(bin_width*len(y_bfor))
    mean_aftr = sum(y_aftr[c_aftr])/len(y_aftr[c_aftr])
    mean_aftr_err = np.sqrt(sum(y_aftr[c_aftr])*bin_width)/(bin_width*len(y_aftr[c_aftr]))

    print("\nTime in 'Before Fit'           = "+str(time_bfor-(mod*bin_width/60.0))+ " min")
    print("Time in 'After Fit'            = "+str(time_aftr)+ " min")
    print("Mixing Time Before 'After Fit' = "+str(mix_t)+ " min"); print(" ")
    
    print("Rn220 Rate Before Injection   = "+str(round(mean_bfor,2))+\
          " +/- "+str(round(mean_bfor_err,2))+" Bq")
    print("Rn220 Rate After Injection    = "+str(round(mean_aftr,2))+\
          " +/- "+str(round(mean_aftr_err,2))+" Bq")
    print("Injected Activity              = "+str(round(mean_aftr-mean_bfor,2))+\
          " +/- "+str(round(np.sqrt((mean_aftr_err**2)+(mean_bfor_err**2)),2))+" Bq"); 

    plt.figure(figsize=(13,4))
    plt.hist(timestamp_min_c, bins=binN, range=[lwrlim,uprlim], \
                         weights=w1, histtype='step', label='All data')
    plt.hist(timestamp_min_c[c_Rn220], bins=binN, range=[lwrlim,uprlim], \
                         weights=w2, histtype='step', label='Rn220 Cut')
    ##############################################################################
    zoomL = min(y1)-min(y1)*0.1; zoomH = max(y1)+max(y1)*0.1
    ##############################################################################
    plt.plot(x_bfor, np.zeros(len(x_bfor))+mean_bfor, color='k', label="Rate Measurements")
    plt.plot(x_aftr, np.zeros(len(x_aftr))+mean_aftr, color='k')
    plt.plot([ts_Inj, ts_Inj],[zoomL, zoomH], color='grey', label="Injection Time")
    plt.xlabel('Minutes since '+files[0]); plt.xlim(lwrlim,uprlim) 
    plt.ylim(zoomL,zoomH); plt.ylabel('Rate [Bq]') 
    plt.legend(loc="upper left"); plt.grid() 

    y_bgnd = y0[low_bfor:high_bfor]
    print("\nDetector Background Rate       = "+str(round(np.mean(y_bgnd),2))+" Hz")
```
